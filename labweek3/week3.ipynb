```python
#1) With given data and using matplotlib library do the following plots:
#a) Histograms, Bar, Pie charts, Scatter and Box.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')
gradeData=pd.read_csv('/content/drive/My Drive/grade_new.csv')
print(gradeData)
```

```python
#histogram
plt.figure(figsize=(10, 6))
plt.hist(gradeData['total marks'], bins=10, color='red')
plt.title('Histogram of Total Marks')
plt.xlabel('Total Marks')
plt.ylabel('Frequency')
plt.show()
```

```python
#bar plot
plt.figure(figsize=(10,6))
plt.bar(gradeData['name'],gradeData['total marks'],color='green')
plt.title('Bar plot of Total Marks')
plt.xlabel('Name')
plt.ylabel('Total Marks')
plt.show()
```

```python
#pie chart
plt.figure(figsize=(10,6))
plt.pie(gradeData['total marks'],labels=gradeData['name'],autopct='%1.1f%%')
plt.title('Pie chart of Total Marks')
plt.axis('equal')
plt.show()
```

```python
#Scatter Plot
plt.figure(figsize=(10,6))
plt.scatter(gradeData['name'],gradeData['total marks'],color='red')
plt.title('Scatter Plot of Total Marks')
plt.xlabel('Name')
plt.ylabel('Total Marks')
plt.show()
```

```python
#Box Plot
plt.figure(figsize=(10,6))
plt.boxplot(gradeData['total marks'],vert=False)
plt.title('Box Plot of Totla Marks')
plt.xlabel('Total Marks')
plt.show()
```

```python
#2
import numpy as np
m1,n1=map(int,input('Enter the row and column of Matrix 1 :').split())
x=np.random.randint(1,10,size=(m1,n1))
matrixA=np.array(x)
m2,n2=map(int,input('Enter the row and column of Matrix 2 :').split())
y=np.random.randint(1,10,size=(m2,n2))
matrixB=np.array(y)
print(f'Matrix 1:-\n{matrixA}')
print(f'Matrix 2:-\n{matrixB}')
```

```python
if matrixA.shape == matrixB.shape:
  matrixSum=matrixA+matrixB
  print(f'Sum of Matrix : \n{matrixSum}')
else:
  print('***Matrix Addiion cannot be performed on uneven dimension***')
```

```python
if n1==m2:
  matrixMul=np.dot(matrixA,matrixB)
  print(f'Product of Matrix :\n {matrixMul}')
else:
  print('***Matrix Multiplication cannot be performed ***')
```

```python
print(f'Transpose of  Sum Matrix:\n{matrixSum.T}')
print(f'Tranpose of Product Matrix:\n{matrixMul.T}')
```

```python
row,col=matrixMul.shape
if row!=col:
  print("***Matrix Inversion cannot be performed because it is not a square matrix***")
else:
  print(f'Matrix Inversion of Product Matrix : \n{ np.linalg.inv(matrixMul)}')
```

```python
#3
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X=np.array([1,2,3,4,5,3,2,4,1,3]).reshape(-1,1)
Y=np.array([71,82,83,94,85,71,82,83,94,85])

model=LinearRegression()
model.fit(X,Y)

#y=mx+c
slope=model.coef_[0]
intercept=model.intercept_
print(f'Linear Regression Equation:y = {slope:.2f}*x + {intercept:.2f}')
```

```python
plt.scatter(X,Y,color='green',label='Data Points')
plt.plot(X,model.predict(X),color='red',label='Line of Best Fit',linewidth=2)
plt.ylabel('MArks Obtained')
plt.xlabel('hours studied')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```
